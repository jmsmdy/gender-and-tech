```python
from bs4 import BeautifulSoup
import requests
response = requests.get('https://www.honeypot.io/women-in-tech-2018/')
html_doc = response.content
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
import pandas as pd
df = pd.read_html(soup.prettify())[0]
```

```python
# df.columns
```

```python
column_data = r"""Total Workforce (Millions)

Female Workforce (Millions)

% Women

% Women Legislators, Senior Officials, and Managers

% Women in Parliament

% Women in Ministerial Positions

Overall Workforce Average Wage (in USD$)

Women's Average Wage (USD$)

Gender Pay Gap (%)

Tech Workforce (Thousands)

% Workforce in Tech

Female Tech Workforce (Thousands)

% Women in Tech

% Difference of Women in Workforce and Women in Tech

Female STEM Graduates (%)

Tech Average Wage (USD$)

Tech Average Wage for Women (USD$)

Gender Pay Gap in Tech (%)

% Difference of Overall Gender Pay Gap and Gender Pay Gap in Tech

Gender Inequality Index (Score, 0 = equality)

Gender Pay Gap 2010

Comparison of Gender Pay Gap From 2010 To 2015
"""
```

```python
columns_pre1 = column_data.split('\n')
```

```python
columns_pre2 = list(filter(lambda x: x != '', columns_pre1))
```

```python
columns = ['#', 'Country'] + columns_pre2
```

```python
df.columns = columns
```
